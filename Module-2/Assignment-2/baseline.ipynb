```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'brand':         ['Toyota', 'honda', 'BMW', ' Ford ', 'toyota', 'Honda', 'BMW', 'ford'],
    'year':          [2015, 2018, 2016, None, 2015, 2018, 2016, 2019],
    'mileage':       ['15000 km', '22000 km', None, '18000 km', '15000 km', '22000 km', None, '31000 km'],
    'fuel_type':     ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', None],
    'selling_price': [500000, 750000, 1200000, None, 500000, 750000, 1200000, 430000]
})
```

1.Brand column has a unique values with same brand, need to fix that.
2.year coumn is in float value
3.missing values are there in fuel type and in targer column selling_price.

```python
df.dropna(subset = ['selling_price'], inplace=True)
```

```python
df['year'].fillna(df['year'].median(),inplace=True)
df['fuel_type'].fillna(df['fuel_type'].mode()[0],inplace=True)
```

```python
df
```

```python
df['brand'] = df['brand'].str.strip().str.lower()
```

```python
df
```

```python
df['mileage'] = df['mileage'].str.extract(r'(\d+)')
df['mileage'] = pd.to_numeric(df['mileage'])
```

```python
df
```

```python
df['mileage'].fillna(df['mileage'].mean(),inplace=True)
```

```python
df.reset_index()
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df.describe()
```

```python
df['year'] = df['year'].astype(int)
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```
