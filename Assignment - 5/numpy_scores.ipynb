```python
import numpy as np

# -----------------------------
# Task 1 — Generate and Inspect
# -----------------------------
np.random.seed(42)

# 5 students × 4 subjects, scores 50–100
scores = np.random.randint(50, 101, (5, 4))
print("Scores:\n", scores)

# 1. Score of 3rd student in 2nd subject
score_3_2 = scores[2, 1]
print("\nScore of 3rd student, 2nd subject:", score_3_2)

# 2. Scores of last 2 students
last_two_students = scores[-2:]
print("\nLast 2 students' scores:\n", last_two_students)

# 3. First 3 students, subjects 2 and 3
subset_scores = np.sum(scores[:3, 1:3])
print("\nFirst 3 students (subjects 2 & 3):\n", subset_scores)

```

```python
# ------------------------------------
# Task 2 — Analyze with Broadcasting
# ------------------------------------

# 1. Column-wise mean (per subject)
col_mean = np.round(np.mean(scores, axis=0), 2)
print("\nColumn-wise mean:", col_mean)

# 2. Add curve with broadcasting
curve = np.array([5, 3, 7, 2])
curved_scores = scores + curve

# Cap scores at 100
curved_scores = np.clip(curved_scores, 0, 100)
print("\nCurved scores:\n", curved_scores)

# 3. Row-wise max (best subject per student)
row_max = np.max(curved_scores, axis=1)
print("\nRow-wise max of curved scores:", row_max)
```

```python
# -------------------------------------------
# Task 3 — Normalize and Identify
# -------------------------------------------

# 1. Min-max normalize row-wise
row_min = curved_scores.min(axis=1, keepdims=True)
row_max_vals = curved_scores.max(axis=1, keepdims=True)
normalized = (curved_scores - row_min) / (row_max_vals - row_min)

print("\nNormalized scores:\n", normalized)

# 2. Identify the single highest value in normalized array
flat_index = np.argmax(normalized)
student_idx, subject_idx = np.unravel_index(flat_index, normalized.shape)

print("\nHighest normalized score at:")
print("Student index:", student_idx)
print("Subject index:", subject_idx)


# 3. Extract all curved scores > 90
above_90 = curved_scores[curved_scores > 90]
print("\nScores above 90:", above_90)
```
