```python
import pandas as pd

data = {
    "EmployeeID": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "Name": ["Sarah", "Michael", "Jennifer", "David", "Lisa", "James", "Emma", "Robert", "Sophia", "William"],
    "Department": ["Engineering", "Sales", "Marketing", "Engineering", "HR", "Sales", "Engineering", "Marketing", "Sales", "HR"],
    "Salary": [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    "Performance Score": [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    "Years of Service": [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}

emp_df = pd.DataFrame(data)
print(emp_df)
```

```python
print(emp_df[['Name','Department','Salary']])
```

```python
print(emp_df.iloc[5])
print(emp_df.loc[7])
```

```python
print(emp_df[emp_df['Department'] == 'Engineering'])
```

```python
print(emp_df[emp_df['Salary'] > 70000])
```

```python
highest_sales = emp_df[(emp_df['Department'] == 'Sales') & (emp_df['Performance Score'] >= 4)]
print(highest_sales)
```

```python
sorted_df = emp_df.sort_values(by='Salary', ascending=False).head(3)
print(sorted_df)
```

```python
sorted_df2 = emp_df.sort_values(['Department','Salary'], ascending=[True,False])
print(sorted_df2)
```
