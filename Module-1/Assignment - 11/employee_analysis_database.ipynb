```python
import sqlite3

conn = sqlite3.connect("employees.db")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute(""" CREATE TABLE IF NOT EXISTS Employees(emp_id INTEGER PRIMARY KEY,name TEXT,department TEXT,salary REAL,years_experience INTEGER,performance_score REAL)""")
```

```python
import random
Employees_data = []

departments = ['Sales', 'HR', 'Marketing', 'Finance', 'Engineering']

for i in range(1,41):
    name = f"Employee{i}"
    department = random.choice(departments)
    salary = round(random.uniform(50000,15000),2)
    years_experience = random.randint(1,15)
    performance_score = round(random.uniform(1.0,5.0),2)

    Employees_data.append((i, name, department, salary, years_experience, performance_score))

print(Employees_data)
```

```python
cursor.executemany(''' INSERT INTO Employees(empy_id, name, department, salary, years_experience, performance_score) VALUES(?,?,?,?,?,?) ''', Employees_data)
```

```python
for row in cursor.execute('SELECT * FROM  Employees'):
    print(row)
```

```python
conn.commit()
```

SQL and Pandas are both powerful tools for working with structured data, but they differ in syntax and typical use cases. SQL is a declarative language designed specifically for querying relational databases. Its syntax uses operators like =, AND, and ORDER BY to filter and sort data. For example, filtering conditions in SQL use AND between clauses, while Pandas uses & between boolean expressions. Similarly, sorting in SQL uses ORDER BY, whereas Pandas uses the .sort_values() method.

Pandas operates inside Python and uses DataFrames to manipulate data. Conditions are written using Python-style comparisons like == and combined using & or |. Pandas is excellent for exploratory analysis, feature engineering, and integrating data manipulation with machine learning workflows.

SQL is typically used when data is stored in databases and you want to retrieve or aggregate information efficiently. It is optimized to process large datasets directly inside the database engine. Pandas is best suited for in-memory analysis when the dataset can fit into system RAM.

For datasets that are too large to fit in memory, SQL is generally the better choice. Databases process data on disk and use indexing, query optimization, and distributed systems to handle very large datasets efficiently without loading everything into memory.
