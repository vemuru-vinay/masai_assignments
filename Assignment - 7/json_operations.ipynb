```python
import json

user_profile = {
    "name": "Vinay",
    "age": 20,
    "email": "vinay@example.com",
    "is_active": True,
    "skills": ["Python", "SQL", "Machine Learning"]
}

json_string = json.dumps(user_profile, indent=2)
print(json_string)
```

```python
import json

api_response = '{"status": "success", "data": {"user_id": 101, "username": "alex99", "score": 87.5}}'

response_dict = json.loads(api_response)

username = response_dict["data"]["username"]
score = response_dict["data"]["score"]

print(username)
print(score)
print(f"User {username} scored {score} points")
```

```python
import json

nested_json = {
    "name": "Priya",
    "address": {
        "city": "Bengaluru",
        "state": "Karnataka",
        "zip": "560001"
    }
}

city = nested_json["address"]["city"]
zipcode = nested_json["address"]["zip"]

print(city)
print(zipcode)

nested_json["address"]["country"] = "India"

print(json.dumps(nested_json, indent=2))
```
